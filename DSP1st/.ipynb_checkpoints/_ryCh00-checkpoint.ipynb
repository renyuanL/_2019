```python
def cosPlot(f=1, A=1):
    #f=  1            # cycle/sec, Hz
    ω= f   *2*np.pi # radian/sec
    ϕ=  0  *2*np.pi
    t=  np.linspace(0,10,1001)

    θ= ω*t + ϕ
    pl.figure()
    pl.plot(t,θ)
    pl.xlabel('t (sec)')
    pl.ylabel('θ (rad)')
    pl.grid()

    x= np.cos(θ) *A
    pl.figure()
    pl.plot(t,x)
    
    pl.xlabel('t (sec)')
    pl.ylabel('x')
    pl.grid()
    
    return x

cosPlot()
```

```python
x= cosPlot(1)
x= cosPlot(2, 2)
x= cosPlot(5, 5)
```

```python
x= cosPlot(10, 10)
```

```python
from IPython.display import Audio
audio = Audio(data= x, rate= 16000)
audio
```

```python
def cosGen(f=1, A=1, T=1, Fs=10000):
    #f=  1            # cycle/sec, Hz
    ω= f   *2*np.pi # radian/sec
    ϕ=  0  *2*np.pi
    t=  np.linspace(0,T,T*Fs+1)
    θ= ω*t + ϕ
    x= np.cos(θ) *A    
    return x

x= cosGen(f= 1000)

audio = Audio(data= x, rate= 10000)
audio
```

```python
x1= cosGen(f= 1000)
x2= cosGen(f= 1500)
x3= cosGen(f= 2000)
x= np.concatenate([x1,x2,x3])

audio = Audio(data= x, rate= 10000)
audio
```

```python
## Major Scale

r= 2**(1/12)
r ### 1.059
majorScale= np.array([0,2,4,5,7,9,11,12])
fM= 1000*(r**majorScale)
fM=np.round(fM)

s= [cosGen(f= f) for f in fM]
x= np.concatenate(s)
audio = Audio(data= x, rate= 10000)
audio
```
