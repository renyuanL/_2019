
# phasor addition

$ 
A_1 cos(\omega t + \phi_1)
+ A_2 cos(\omega t + \phi_2)
= A cos(\omega t + \phi)    .... (1)
$

$ 
A_1 sin(\omega t + \phi_1)
+ A_2 sin(\omega t + \phi_2)
= A sin(\omega t + \phi)    .... (2)
$

$
(1) + j*(2) \\
\Rightarrow \\
A_1 e^{j(\omega t + \phi_1)}
+ A_2 e^{j(\omega t + \phi_2)}
= A e^{j(\omega t + \phi)}
$

$
\Rightarrow  \\
  A_1 e^{j(\omega t)}  e^{j\phi_1}
+ A_2 e^{j(\omega t)}  e^{j\phi_2}
= A   e^{j(\omega t)}  e^{j(\phi)}
$






$
\Rightarrow \\
  A_1   e^{j\phi_1}
+ A_2   e^{j\phi_2}
= A     e^{j\phi}
$


$
\Rightarrow \\
  A_1 cos(\phi_1) + j A_1 sin(\phi_1) \\
+ A_2 cos(\phi_2) + j A_2 sin(\phi_2) \\
= A e^{j\phi} \\
= A [ cos(\phi) + j sin(\phi) ] \\
= X + j Y
$

$
\Rightarrow \\
X= A cos(\phi) = A_1 cos(\phi_1) + A_2 cos(\phi_2) \\
Y= A sin(\phi) = A_1 sin(\phi_1) + A_2 sin(\phi_2)
$

$
\Rightarrow \\
A = \sqrt{X^2+Y^2}
$

$
tan(\phi) = \frac{Y}{X} \\
\Rightarrow   \\
\phi= tan^{-1}(\frac{Y}{X})
$

$
\phi= tan^{-1}(\frac{Y}{X})
$

## PHASOR ADDITION RULE

```python
import numpy as np

π=  np.pi

A1= 1.7
ϕ1= 70*π/180
A2= 1.9
ϕ2= 200*π/180

ω= 2*π*10
t= np.linspace(0,1,1001)
z1= A1*np.exp(1j*(ω*t+ϕ1))
z2= A2*np.exp(1j*(ω*t+ϕ2))
z3= z1+z2

# z3= A *np.exp(1j *(ω*t +ϕ))
# A *np.exp(1j*ϕ) == A1 *np.exp(1j*ϕ1) + A2 *np.exp(1j*ϕ2)

X= A1 * np.cos(ϕ1) + A2 * np.cos(ϕ2) 
Y= A1 * np.sin(ϕ1) + A2 * np.sin(ϕ2)
A= np.sqrt(X**2 +Y**2)
ϕ= np.arctan2(Y,X)

_z3= A *np.exp(1j *(ω*t +ϕ))
A, ϕ
```

```python
z3 == _z3  ## array([False, False, False, ..., False])
# they are not the same exactly, 
# because they are truncated numbers...
```

```python
np.abs(z3-_z3).sum()  # == 3.4737011292329153e-12

# the difference between them is very small, 
# so we can look them upon as the same
```

```python
import matplotlib.pylab as plt

plt.plot(z3.real)
plt.plot(_z3.real,'--')
plt.show()

plt.figure()
plt.plot(z3.imag)
plt.plot(_z3.imag,'--')
plt.show()
```
