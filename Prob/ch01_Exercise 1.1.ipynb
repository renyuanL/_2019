# Ex 1.5



```python
cGroup= [7, 3, -4, 14, 2,  5, 22, -7, 9, 5]
tGroup= [-6, 5, 9, 4, 4,   12, 37, 5, 3, 3]

group= np.array([cGroup, tGroup])
group
```

```python
group.size, group.shape
```

```python
aRange= (group.min(), group.max())
nBin= group.max() - group.min() +1

plt.figure(0)
plt.hist(group[0], range= aRange, bins= nBin)
plt.grid()

plt.figure(1)
plt.hist(group[1], range= aRange, bins= nBin)
plt.grid()
```

```python
# help(plt.hist)
#  range : tuple or None, optional
#          The lower and upper range of the bins. Lower and upper outliers
#          are ignored. If not provided, *range* is ``(x.min(), x.max())``.
#          Range has no effect if *bins* is a sequence.
```
